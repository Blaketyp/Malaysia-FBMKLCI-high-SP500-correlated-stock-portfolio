```python
import pandas as pd
import pandas_datareader as pdr
import datetime as dt
import numpy as np
```

```python
tickers = ['6888.KL', '1023.KL', '7277.KL', '6947.KL', '3182.KL', '4715.KL', '5168.KL', '5819.KL', '1082.KL', '5225.KL', '1961.KL', '2445.KL', '1155.KL', '6012.KL', '3816.KL', '4707.KL', '5183.KL', '5681.KL', '6033.KL', '4065.KL', '8869.KL', '1295.KL', '1066.KL', '4197.KL', '5285.KL', '4863.KL', '5347.KL', '7113.KL']
start = dt.datetime(2020, 1, 1)
```

```python
data = pdr.get_data_yahoo(tickers, start)
data = data['Adj Close']
```

```python
log_returns = np.log(data/data.shift())
```

```python
log_returns.corr()
```

```python
sp500 = pdr.get_data_yahoo("^GSPC", start)
 
log_returns['SP500'] = np.log(sp500['Adj Close']/sp500['Adj Close'].shift())
 
log_returns.corr()
```

```python
finalcorr = log_returns.corr()

sp500corr = finalcorr.iloc[:,-1:]

highest_corr = sp500corr.sort_values("SP500",ascending = False).head(11)
highest_corr
```
